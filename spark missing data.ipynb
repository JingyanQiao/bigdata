```
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('miss').getOrCreate()
```

```
path = '/FileStore/tables/ContainsNull.csv'
df = spark.read.csv(path, inferSchema = True, header = True)
```

```
df.show()
```

```
df.printSchema()
```

```
# drop missing data
df.na.drop().show() # will drop any row that include null value
```

```
df.na.drop(thresh = 2).show() # only keep rows that include >=2 not null values
```

```
df.na.drop(how = 'all').show() # only drop the row that all null values
```

```
df.na.drop(subset = ['Sales']).show() # drop rows when sales is null
```

```
df.na.fill('FILL VALUES').show() # fill the string to any column that is string type
```

```
df.na.fill(0).show() 
```

```
df.na.fill(0, subset = ['Sales']).show() # use subset to specify the column you want to fill
```
